```python
import seaborn as sns
import pandas as pd

# 예시 데이터셋 로드 (titanic 데이터셋)
titanic = sns.load_dataset('titanic')

# 숫자형 데이터만 선택
numeric_titanic = titanic.select_dtypes(include=['float64', 'int64'])

# 상관 계수 계산
titanic_corr = numeric_titanic.corr(method='pearson')

# 상관 계수 출력
print(titanic_corr)
```

```python
titanic['survived'].corr(titanic['adult_male'])
titanic['survived'].corr(titanic['fare'])
```

```python
sns.pairplot(titanic,hue ='survived') # 산점도 (scatter plot) 행렬
plt.show()
```

```python
sns.catplot(x='pclass',y='survived',hue='sex',data =titanic, kind='point')
plt.show()
```

```python
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
    
titanic['age2'] = titanic['age'].apply(category_age)
titanic['sex'] = titanic['sex'].map({'male':1, 'female':0})
titanic['family'] = titanic['sibsp'] + titanic['parch'] +1
titanic.to_csv('C:/Users/it/.ipython/titanic.csv', sep=';', index=False)
heatmap_data = titanic[['survived', 'sex', 'age2', 'family', 'fare', 'pclass']]
colormap = plt.cm.RdBu
sns.heatmap(heatmap_data.astype(float).corr(), linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True, annot_kws={"size": 10})
plt.show()
```
